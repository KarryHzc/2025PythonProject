```python

from matplotlib import pyplot as plt  # 导入Matplotlib的pyplot模块，通常别名为plt
x = range(1,8) # x轴的位置 (数据为[1, 2, 3, 4, 5, 6, 7])
y = [17, 17, 18, 15, 11, 11, 13] # y轴对应的数值
# 传入x和y, 通过plot画折线图
# plt.plot(x, y, marker='o') # 绘制折线图，并指定数据点标记为圆形
plt.plot(x, y, 
         marker='D',         # 菱形标记
         markersize=5,      # 大小
         markerfacecolor='yellow', # 填充黄色
         markeredgecolor='black', # 边缘黑色
         markeredgewidth=2,  # 边缘宽度
         linestyle='--',     # 虚线连接
         color='green'       # 线条颜色
        )
plt.show() # 显示图形
''' 折点形状选择:
================ ===============================
character description
================ ===============================
``'-'`` solid line style
``'--'`` dashed line style
``'-.'`` dash-dot line style
``':'`` dotted line style
``'.'`` point marker
``','`` pixel marker
``'o'`` circle marker
``'v'`` triangle_down marker
``'^'`` triangle_up marker
``'<'`` triangle_left marker
``'>'`` triangle_right marker
``'1'`` tri_down marker
``'2'`` tri_up marker
``'3'`` tri_left marker
``'4'`` tri_right marker
``'s'`` square marker
``'p'`` pentagon marker
``' '`` star marker
``'h'`` hexagon1 marker
``'H'`` hexagon2 marker
``'+'`` plus marker
``'x'`` x marker
``'D'`` diamond marker
``'d'`` thin_diamond marker
``'|'`` vline marker
``'_'`` hline marker
'''
```

```python
# 7.8扩展(一图多个坐标系子图)

import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1, 100) #划分子图
# NumPy 中用于生成等差数列的函数
fig,axes=plt.subplots(2,2)
'''
这个函数会返回一个元组 (fig, ax)
fig：是一个Figure对象。你可以把它想象成整个画布或画板，是所有子图的顶级容器
ax（或你代码中的 axes）：通常是一个 NumPy 数组。这个数组里的每个元素都是一个 Axes对象，代表画布上的一个子图（可以绘图的区域）。对于 (2, 2)的网格，ax就是一个 2x2 的数组

元组解包：Python 的元组解包语法 a, b = (value_a, value_b)允许我们将返回的元组直接拆开，并分别赋值给两个变量。
'''
ax1=axes[0,0]
ax2=axes[0,1]
ax3=axes[1,0]
ax4=axes[1,1]
#fig=plt.figure(figsize=(20,10),dpi=80) #作图 1
ax1.plot(x, x) # 作 图 2
ax2.plot(x, -x) #作图 3
ax3.plot(x, x**2)
# ax3.grid(color='r', linestyle='--', linewidth=1,alpha=0.3) #作图 4
ax4.plot(x, np.log(x))
plt.show()
```

```python
# add_subplot 方法----给 figure 新增子图
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,100)
#新建 figure 对象
fig=plt.figure(figsize=(20,10),dpi=80) #新建子图 1
ax1=fig.add_subplot(2,2,1)
# ax = fig.add_subplot(nrows, ncols, index, **kwargs)中index索引从1开始
ax1.plot(x, x)
#新建子图 2
ax3=fig.add_subplot(2,2,2)
ax3.plot(x, x**2)
ax3.grid(color='r', linestyle='--', linewidth=1,alpha=0.3)
# 新 建 子 图 3
ax4=fig.add_subplot(2,2,3)
ax4.plot(x, np.log(x))
plt.show()
```

```python
# 9 直方图
'''假设你获取到了 2019 年内地电影票房前 20 的电影（列表 a)和电影票房数据（列表 b)，请展示该数
据
a = ['流浪地球','疯狂的外星人','飞驰人生','大黄蜂','熊出没·原始时代','新喜剧之王']
b = [38.13,19.85,14.89,11.36,6.47,5.93]'''
from matplotlib import pyplot as plt
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\STSONG.TTF',size=16)
a = ['流浪地球','疯狂的外星人','飞驰人生','大黄蜂','熊出没·原始时代','新喜剧之王']
b = [38.13,19.85,14.89,11.36,6.47,5.93]
plt.figure(figsize=(20,8),dpi=80)
# 绘制条形图的方法
'''
width=0.3 条形的宽度
'''
# 绘制柱状图
# plt.bar(x, height, width, color)
# x: 条形的位置
# height: 条形的高度
# width: 条形的宽度
# color: 条形的颜色
rects = plt.bar(range(len(a)),b,width=0.3,color='r')
'''
plt.xticks(ticks, labels, fontproperties, rotation) 方法用于设置 x 轴刻度
ticks：表示 x 轴刻度的位置
labels：表示 x 轴刻度的标签
fontproperties：设置字体
rotation：设置旋转角度
'''
plt.xticks(range(len(a)),a,fontproperties=my_font,rotation=45)
# 在条形图上加标注(水平居中)
'''
plt.text(x, y, s, ha) 方法用于在图上添加文字
x：表示文字的 x 坐标
y：表示文字的 y 坐标
s：表示文字内容
ha：表示水平对齐方式，center 表示水平居中
va：表示垂直对齐方式，center 表示垂直居中
'''
for rect in rects:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height+0.3, str(height),ha="center")
plt.show()
```

```python
# 10 饼图
import matplotlib.pyplot as plt
import matplotlib
label_list = ["第一部分", "第二部分", "第三部分"] # 各部分标签
size = [55, 35, 10] # 各部分大小
color = ["red", "green", "blue"] # 各部分颜色
explode = [0, 0.05, 0] # 各部分突出值
""" 绘 制 饼 图
explode： 设 置 各 部 分 突 出
label: 设 置 各 部 分 标 签
labeldistance:设置标签文本距圆心位置，1.1 表示 1.1 倍半径
autopct： 设 置 圆 里 面 文 本
shadow：设置是否有阴影
startangle：起始角度，默认从 0 开始逆时针转
pctdistance：设置圆内文本距圆心距离返回值
l_text：圆内部文本，
matplotlib.text.Text object
p_text：圆外部文本"""
matplotlib.rcParams['font.sans-serif']=['SimHei']
patches, l_text, p_text = plt.pie(size,explode=explode, colors=color, labels=label_list,
labeldistance=1.1, autopct="%1.1f%%", shadow=False,
startangle=90, pctdistance=0.6)

plt.axis("equal") # 设置横轴和纵轴大小相等，这样饼才是圆的
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib

# 设置全局字体以正确显示中文
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 数据准备
label_list = ["第一部分", "第二部分", "第三部分"]
size = [55, 35, 10]
color = ["#ff9999", "#66b3ff", "#99ff99"]  # 使用更柔和的十六进制颜色
explode = (0, 0.05, 0)  # 突出第二部分

# 创建图形和轴对象
fig, ax = plt.subplots(figsize=(8, 6))

# 绘制饼图 (使用更丰富的参数)
wedges, texts, autotexts = ax.pie(
    size,
    explode=explode,
    colors=color,
    labels=label_list,
    autopct='%1.1f%%',       # 显示百分比，保留一位小数
    startangle=90,           # 从90度（正北）开始绘制
    shadow=False,            # 无阴影
    pctdistance=0.85,       # 百分比标签放在扇形内部靠外位置
    wedgeprops={'edgecolor': 'w', 'linewidth': 2},  # 白色边框，线宽为2
    textprops={'fontsize': 10, 'color': 'darkblue'} # 统一设置文本样式
)

# 进一步美化：将百分比文本设置为白色并加粗
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# 设置标题
plt.title('示例饼图', fontsize=14, fontweight='bold')

# 确保饼图是圆形
ax.axis('equal')

plt.show()
```
